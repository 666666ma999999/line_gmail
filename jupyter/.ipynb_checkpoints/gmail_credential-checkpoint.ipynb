```python

import gspread
from oauth2client.service_account import ServiceAccountCredentials
from google.oauth2 import service_account
import csv
import pandas as pd


# 設定
json_spread = 'linesales20230706-02-25146e4e9b26.json'
file_name = 'LineSales'
sheet_name1 = 'sheet1'
sheet_name2 = 'sheet2'

scopes_spreadsheet = [
    "https://spreadsheets.google.com/feeds",
    "https://www.googleapis.com/auth/drive" ,
    "https://www.googleapis.com/auth/drive.file" ,
    "https://www.googleapis.com/auth/spreadsheets"
    ]


#setting
#credentials = ServiceAccountCredentials.from_json_keyfile_name(json_file, scope).with_subject('uranaifree123@gmail.com')
credentials = service_account.Credentials.from_service_account_file(json_spread, scopes=scopes_spreadsheet)

gc = gspread.authorize(credentials)
sh = gc.open(file_name)

# 変数設定（日付）
import datetime
from dateutil.relativedelta import relativedelta

#before_date = datetime.datetime.now()
before_date = datetime.datetime.now()+ relativedelta(days=-0)
after_date = before_date + relativedelta(days=-1)
print(before_date , after_date)
```

```python
import base64
import json
import io
import csv
import os.path
from googleapiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow
from google.auth.transport.requests import Request
import base64
from email.mime.text import MIMEText
from apiclient import errors
import logging
from gspread_dataframe import set_with_dataframe

#from docopt import docopt
#from gmail_credential import get_credential

import sys
if sys.version_info[0] < 3: 
    from StringIO import StringIO
else:
    from io import StringIO

# Gmail APIのクライアントを作成
service = build('gmail', 'v1', credentials=creds, cache_discovery=False)



# メールの検索クエリを設定
attachment = 'has:attachment'
target_email = 'dev-line-fortune@linecorp.com'
target_date = f'before:{before_date.strftime("%Y/%m/%d")} after:{after_date.strftime("%Y/%m/%d")}'

#print(query)

# Gmailのメッセージを検索
query = f'from:{target_email} {target_date} {attachment}'
response = service.users().messages().list(userId='me', q=query).execute() #, maxResults=500).execute()
#print(response)

messages = response.get('messages', [])


sales_daily = []
for message in messages[:]:
    msg = service.users().messages().get(userId='me', id=message['id']).execute()
    payload = msg['payload']

        #print(payload)

    # 添付ファイルの処理
    if 'parts' in payload:
        parts = payload['parts']
        for part in parts:
            if 'filename' in part:
                filename = part['filename']
                body = part['body']

                # 添付ファイルのデータを取得
                if 'attachmentId' in body:
                    attachment = service.users().messages().attachments().get(
                            userId='me',
                            messageId=message['id'],
                            id=body['attachmentId']
                    ).execute()

                    import csv
                    data = attachment['data']
                    csv_data = base64.urlsafe_b64decode(data).decode('utf-8')
                    df = pd.read_csv(StringIO(csv_data))
                        
                        #file_data = base64.urlsafe_b64decode(data.encode('utf8'))

                    web_sales = [df['service_name'][0] , 'web'  , df['web_paid_amount'].sum()]
                    ios_sales = [df['service_name'][0] , 'ios'  , df['ios_paid_cost'].sum()]
                    android_sales = [df['service_name'][0] , 'android'  , df['android_paid_cost'].sum()]
                    
                    sales_daily.append(web_sales)
                    sales_daily.append(ios_sales)
                    sales_daily.append(android_sales)
                             
                    #print(sales)
                
                   #print(df[['service_name', 'android_paid_amount','ios_paid_amount','web_paid_amount', 'date']])

df_sales = pd.DataFrame(sales_daily , columns=['title', 'pf', after_date.strftime("%Y/%m/%d")]) 
#df_sales.to_csv('sales_0623.csv' , encoding='cp932' , index=False)

print(df_sales)
```

```python
#取得ファイルをシートに保存
newday_wks = sh.add_worksheet(title=after_date.strftime("%Y/%m/%d"), rows='500', cols='3')
set_with_dataframe(newday_wks, df_sales)
#同名シートがあるとエラーを起こすので注意が必要
```

```python
#元ファイルを読み込む
wks = sh.worksheet(sheet_name1)
df_origin = pd.DataFrame(wks.get_all_records())

# sheet1の上書き保存
df_overwrite= pd.merge(df_origin , df_sales , on=['title' , 'pf'] , how='outer')

#nan対策
df_overwrite = df_overwrite.fillna(0)

#小数点以下対策
for c in df_overwrite.columns:
    #print(c, df_overwrite[c].dtype)
    if df_overwrite[c].dtype == 'float64':
        df_overwrite[c] = df_overwrite[c].astype('int64')

set_with_dataframe(wks, df_overwrite)
```
